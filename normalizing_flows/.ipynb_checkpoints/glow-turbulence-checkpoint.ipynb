## Glow normalizing flows

This notebook provides a simple example of training Glow on turbulent flow data.

Note that, for the default architecture settings, you *will* need a GPU in order to run this code.

The size of the model can be adjusted via the `num_layers` and `depth` arguments, as well as the `coupling_nn_ctor` argument, which builds the deep neural networks used in the coupling layers.

```python
def spectrum(image):
    n, _ = image.shape
    fourier_image = np.fft.fftn(image) / n**2
    fourier_amplitudes = np.abs(fourier_image)**2
    
    kmin = 1.0
    kmax = kmin + n//2+1
    
    kbins = np.arange(kmin, kmax, kmin)
    N = len(kbins)

    kfreq = np.fft.fftfreq(n) * n
    
    kfreq2D = np.meshgrid(kfreq, kfreq)
    k = np.sqrt(kfreq2D[0]**2 + kfreq2D[1]**2)
    
    whichbin = np.digitize(k.flat, kbins)
    ncount = np.bincount(whichbin)
    
    E_k = np.zeros(len(ncount)-1)
    
    for n in range(1,len(ncount)):
        E_k[n-1] = np.sum(fourier_amplitudes.flat[whichbin==n])

    k = 0.5*(kbins[0:N-1] + kbins[1:N])
    E_k = E_k[1:N]
    
    return k[:-1], E_k[:-1]
    
    #kbins = np.arange(0.5, n//2 + 1, 1.)
    #kvals = 0.5 * (kbins[1:] + kbins[:-1])
    #Abins, _, _ = stats.binned_statistic(knrm, fourier_amplitudes,
    #                                     statistic = "sum",
    #                                     bins = kbins)
    #
    #return kvals, Abins
```

```python
def readH5(f):
    f = f.numpy()
    file = f[0]
    idx =  f[1]
    with h5py.File(f, 'r') as fi:
        X = fi['HR/ux'][:,:,:].reshape(256,256,256,1)
        m = tf.math.reduce_mean(X)
        s = tf.math.reduce_std(X)

        X = X[:128,:128,:]
        #X = tf.math.log(X)
        X = (X-m)/s
        X = tf.cast(X, tf.float32)
        Y = tf.zeros([1], tf.float32)
        return X, Y


def readH5_wrapper(sliceList):
    # Assuming your data and labels are float32
    # Your input is parse_function, who arg is filename, and you get X and y as output
    # whose datatypes are indicated by the tuple argument
    features, labels = tf.py_function(readH5, [sliceList], (tf.float32, tf.float32))
    return features#, labels

fList_t  = glob.glob(dset_dir+'/train/*/processed_data/snapshot.h5')
idxList = [i for i in range(256)]
sliceList_t = [[a, b] for a in fList_t for b in idxList] 
sliceList_t = np.array(sliceList_t)

batch_size = 8
# Create dataset of filenames.
X_train_ds = tf.data.Dataset.from_tensor_slices(sliceList_t)
X_train_ds = X_train_ds.map(readH5_wrapper)
X_train_ds = X_train_ds.batch(8)
```
